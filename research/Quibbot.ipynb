```python
from transformers import AutoModelForSeq2SeqLM, AutoTokenizer

device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
model_ckpt = "google/pegasus-cnn_dailymail"

tokenizer = AutoTokenizer.from_pretrained(model_ckpt)

model_pegasus = AutoModelForSeq2SeqLM.from_pretrained(model_ckpt).to(device)
```

```python
dataset_samsum = load_from_disk('samsum_dataset')
dataset_samsum
```

```python
def convert_examples_to_features(example_batch):
  input_codings = tokenizer(example_batch['dialogue'], max_length=1024, truncation=True)
  with tokenizer.as_target_tokenizer():
    target_encodings = tokenizer(example_batch['summary'], max_length=128, truncation=True)
  return {
      'input_ids': input_codings['input_ids'],
      'attention_mask': input_codings['attention_mask'],
      'labels': target_encodings['input_ids']
  }
```

```python
dataset_samsum_pt = dataset_samsum.map(convert_examples_to_features, batched = True)
```

```python
dataset_samsum_pt['train']
```

```python
from transformers import DataCollatorForSeq2Seq

seq2seq_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model_pegasus)
```

```python
import transformers
print(transformers.__version__)
```

```python
from transformers import TrainingArguments, Trainer

trainer_args = TrainingArguments(
    logging_dir='./logs',
    output_dir='pegasus-samsum',              # thư mục lưu mô hình
    num_train_epochs=3,                       # số epoch huấn luyện
    warmup_steps=500,                         # số bước khởi động trước khi learning rate giảm
    per_device_train_batch_size=1,            # batch size cho mỗi thiết bị khi train
    per_device_eval_batch_size=1,             # batch size khi evaluate
    weight_decay=0.01,                        # hệ số giảm trọng số để tránh overfitting
    logging_steps=10,                         # mỗi 10 step thì ghi log một lần
    eval_strategy='steps',              # đánh giá theo số bước
    eval_steps=100,                           # mỗi 500 bước sẽ đánh giá
    save_steps=1e6,                           # lưu mô hình mỗi 1 triệu bước (rất lâu, gần như không lưu giữa chừng)
    gradient_accumulation_steps=16,            # tích luỹ gradient trong 16 bước trước khi cập nhật trọng số
    save_strategy='epoch'
)
```

```python
trainer = Trainer(model=model_pegasus, args=trainer_args, tokenizer=tokenizer, data_collator=seq2seq_data_collator,
                  train_dataset=dataset_samsum_pt['test'], eval_dataset=dataset_samsum_pt['validation'])
```

```python
trainer.train()
```

```python
def generate_batch_sized_chunks(list_of_elements, batch_size):
    """Split the dataset into smaller batches that we can process simultaneously.
    Yield successive batch-sized chunks from list_of_elements."""
    for i in range(0, len(list_of_elements), batch_size):
        yield list_of_elements[i : i + batch_size]


def calculate_metric_on_test_ds(dataset, metric, model, tokenizer,
                                batch_size=16, device=device,
                                column_text="article",
                                column_summary="highlights"):
    article_batches = list(generate_batch_sized_chunks(dataset[column_text], batch_size))
    target_batches = list(generate_batch_sized_chunks(dataset[column_summary], batch_size))

    for article_batch, target_batch in tqdm(
        zip(article_batches, target_batches), total=len(article_batches)
    ):
        inputs = tokenizer(article_batch, max_length=1024, truncation=True,
                           padding="max_length", return_tensors="pt")
        summaries = model.generate(input_ids=inputs["input_ids"].to(device),
                                   attention_mask=inputs["attention_mask"].to(device),
                                   length_penalty=0.8, num_beams=8, max_length=128)

        decoded_summaries = [tokenizer.decode(s, skip_special_tokens=True,
                                              clean_up_tokenization_spaces=True)
                             for s in summaries]
        metric.add_batch(predictions=decoded_summaries, references=target_batch)

    # Finally compute and return ROUGE scores
    score = metric.compute()
    return score
```

```python
from evaluate import load
rouge_names = ["rouge1", "rouge2", "rougeL", "rougeLsum"]
rouge_metric = load('rouge')
```

```python
score = calculate_metric_on_test_ds(
    dataset_samsum['test'][0:10], rouge_metric, trainer.model, tokenizer, batch_size = 2, column_text = 'dialogue', column_summary= 'summary'
)
rouge_dict = {rn: round(score[rn], 4) for rn in rouge_names}
pd.DataFrame(rouge_dict, index = [f'pegasus'])
```
